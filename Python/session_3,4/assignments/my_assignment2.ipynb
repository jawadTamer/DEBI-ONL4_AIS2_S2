```python
import os
import random

dir ="C:/Users/jawad/OneDrive/Desktop/DEBI-ONL4_AIS2_S2/Python/session_3,4/assignments/assignment_2"
def manage_folder(folder_path, num_files):
    """
    Manage a folder by creating, checking, and deleting files.
    
    Args:
        folder_path: Path to the folder to manage
        num_files: Number of files to create (default: 10)
    """
    if os.path.exists(folder_path):
        print(f"The folder '{folder_path}' already exists.")
    else:
        os.makedirs(folder_path)
        print(f"The folder '{folder_path}' has been created.")
    for i in range(num_files):
        inner_files=os.path.join(folder_path, f"file_{i+1}")
        if not os.path.exists(inner_files):
            os.makedirs(inner_files)
            print(f"Created folder: {inner_files}")
            
    items = os.listdir(folder_path)
    print(f"\nTotal number of items : {len(items)}")
    
    num_to_delete = len(items) // 2
    items_to_delete = random.sample(items, num_to_delete)
    for item in items_to_delete:
        item_path = os.path.join(folder_path, item)
        if os.path.isdir(item_path):
            os.rmdir(item_path)
            print(f"Deleted folder: {item_path}")
        else:
            os.remove(item_path)
            print(f"Deleted file: {item_path}")
            
    items = os.listdir(folder_path)
    print(f"\nTotal number of items : {len(items)}")        

manage_folder(dir, 10)    
```
