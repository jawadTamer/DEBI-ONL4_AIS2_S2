```python
"""
    PalindromeChecker class checks if a given string is a palindrome.

    Attributes:
        text (str): The string to be checked.
"""
class PalindromeChecker:

    def __init__(self, text):
        """
        Initializes PalindromeChecker with the provided string.

        Args:
            text (str): The string to check.
        """
        self.text = text

    def is_palindrome(self):
        """
        Checks if the stored string is a palindrome.

        Returns:
            str: 'Palindrome' if the string is a palindrome, otherwise 'Not a Palindrome'.
        """
        if self.text == self.text[::-1]:
            return "Palindrome"
        else:
            return "Not a Palindrome"


user_input = input("Enter a string: ").strip()
checker = PalindromeChecker(user_input)
print(checker.is_palindrome())
```

```python
"""
    FactorialCalculator calculates the factorial of a number.

    Attributes:
        number (int): The number to calculate the factorial for.
"""
class FactorialCalculator:

    def __init__(self, number):
        """
        Initializes FactorialCalculator with the provided number.

        Args:
            number (int): The number to calculate the factorial for.
        """
        self.number = number

    def factorial(self, n=None):
        """
        Calculates the factorial recursively.

        Returns:
            int: The factorial of the number.
        """
        if n is None:
            n = self.number
        if n == 0 or n == 1:
            return 1
        else:
            return n * self.factorial(n - 1)


fact_calc = FactorialCalculator(5)
print(fact_calc.factorial())
```

```python
"""
    SquaresListGenerator generates a list of squares from 1 to n.

    Attributes:
        n (int): The upper limit.
"""
class SquaresListGenerator:

    def __init__(self, n):
        """
        Initializes SquaresListGenerator with the upper limit n.

        Args:
            n (int): The upper limit.
        """
        self.n = n

    def generate(self):
        """
        Generates the list of squares from 1 to n.

        Returns:
            list: List of squares.
        """
        return [i ** 2 for i in range(1, self.n + 1)]


lst_of_squares = SquaresListGenerator(10)
print(lst_of_squares.generate())
```

```python
""" 
    EvenSumCalculator calculates the sum of even numbers in a list.

    Attributes:
        numbers (list): The list of numbers.

"""
class EvenSumCalculator:

    def __init__(self, numbers):
        """
        Initializes EvenSumCalculator with the list of numbers.

        Args:
            numbers (list): The list of numbers.
        """
        self.numbers = numbers

    def sum_even(self):
        """
        Calculates the sum of even numbers in the list.

        Returns:
            int: The sum of even numbers.
        """
        return sum(num for num in self.numbers if num % 2 == 0)


even_sum = EvenSumCalculator([1, 2, 3, 4, 5, 6])
print(even_sum.sum_even())
```

```python
"""
ReverseList class reverses a given list.
Attributes:
    lst (list): The list to be reversed.
"""
class reverseList:
    
    def __init__(self, lst):
        """
        Initializes reverseList with the provided list.
        Args:
            lst (list): The list to reverse.
        """
        self.lst = lst

    def reverse(self):
        """
        Reverses the stored list.

        Returns:
            list: The reversed list.
        """
        return self.lst[::-1]

reversed = reverseList([1, 2, 3, 4, 5])
print(reversed.reverse())    
```

```python
"""
CalculateMedian class calculates the median of a list of numbers.
Attributes:
    numbers (list): The list of numbers.
"""
class calculateMedian:
    
    def __init__(self, numbers):
        """
        Initializes calculateMedian with the list of numbers.
        Args:
            numbers (list): The list of numbers.
        """
        self.numbers = sorted(numbers)

    def median(self):
        """
        Calculates the median of the list.

        Returns:
            float: The median value.
        """
        n = len(self.numbers)
        mid = n // 2
        if n % 2 == 0:
            return (self.numbers[mid - 1] + self.numbers[mid]) / 2
        else:
            return self.numbers[mid]


list_for_median = calculateMedian([5,2,7,3,1])     
print(list_for_median.median())   
```

```python
"""
FindMode class finds the mode(s) of a list of numbers.
Attributes:
    numbers (list): The list of numbers.
"""
class findMode:
    
    def __init__(self, numbers):
        """
        Initializes findMode with the list of numbers.
        Args:
            numbers (list): The list of numbers.
        """
        self.numbers = numbers

    def mode(self):
        """"
        Finds the mode of the list.
        Returns:
            list: The mode value(s).
        """
        
        frequency = {}
        for num in self.numbers:
            frequency[num] = frequency.get(num, 0) + 1
        max_freq = max(frequency.values())
        modes = [num for num, freq in frequency.items() if freq == max_freq]
        return modes
    
mode_finder = findMode([5,2,7,3,2,1,5])
print(mode_finder.mode())    
```

```python
"""
ConvertGrade class converts numerical grades to letter grades.
Attributes:
    grades (list): The list of numerical grades.
"""
class convertGrade:
    
    def __init__(self, marks):
        """
        Initializes convertGrade with the list of numerical grades.
        Args:
            grades (list): The list of numerical grades.
        """
        self.marks = marks
    def convert(self):
        """Converts numerical grades to letter grades.
        Returns:
            list: The list of letter grades.
        """
        mark = int(self.marks)
        if 85 <= mark <= 100:
            return 'A'
        elif 75 <= mark < 85:
            return 'B'
        elif 65 <= mark < 75:
            return 'C'
        elif 50 <= mark < 65:
            return 'D'
        elif 0 <= mark < 50:
            return 'F'
        else:
            return 'Invalid mark'

marks_converter = convertGrade(75)
print(marks_converter.convert())        
```
