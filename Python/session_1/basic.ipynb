```python
# Task_1
email = "Amit_ml@gmail.edu"

#!1.Input Validation
if email.count("@") != 1:
    print("Invalid email")
else:
    at_index = email.find("@")
    after_at = email[at_index + 1:]
    if "." not in after_at:
        print("Invalid email")
    else:
         
        #! 2.Extract Username
        username = email[:at_index]
        print("Username:", username)
        
        #! 3.Extract Domain
        last_dot = email.rfind(".")
        domain = email[at_index + 1:last_dot]
        print("Domain:", domain)
        
        #! 4.Check for Domain Ending
        if email.endswith(".com"):
            print("Commercial Domain")
        elif email.endswith(".edu"):
            print("Educational Domain")
        else:
            print("Other Domain")
```
