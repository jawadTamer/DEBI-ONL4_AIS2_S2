```python
# Task_1
email = "Amit_ml@gmail.edu"

#!1.Input Validation
if email.count("@") != 1:
    print("Invalid email")
else:
    at_index = email.find("@")
    after_at = email[at_index + 1:]
    if "." not in after_at:
        print("Invalid email")
    else:
         
        #! 2.Extract Username
        username = email[:at_index]
        print("Username:", username)
        
        #! 3.Extract Domain
        last_dot = email.rfind(".")
        domain = email[at_index + 1:last_dot]
        print("Domain:", domain)
        
        #! 4.Check for Domain Ending
        if email.endswith(".com"):
            print("Commercial Domain")
        elif email.endswith(".edu"):
            print("Educational Domain")
        else:
            print("Other Domain")
```

```python
# Task_2
message="###!!@mocleW EPGTQ!!!6789 "

start=0
while not message[start].isalpha():
    start+=1
    
end=len(message)-1 
while not message[end].isalpha():
    end-=1

extracted_message=message[start:end+1]

word1, word2 = extracted_message.split()
reversed_word1 = word1[::-1]

decoded_message = reversed_word1 + " " + word2
print(f"Decoded Message: {decoded_message}")    
```

```python
# Task_3
message="&&&**$gnirtS PLIO!!@1234"

start=0
while not message[start].isalpha():
    start+=1
    
end=len(message)-1 
while not message[end].isalpha():
    end-=1
    
extracted_message=message[start:end+1]

word1, word2 = extracted_message.split()
reversed_word1 = word1[::-1]

shifted_vowels = {
    "I": "E",
    "O": "U",
}

shifted_word2 = ""
for char in word2:
    if char in shifted_vowels:
        shifted_word2 += shifted_vowels[char]
    else:
        shifted_word2 += char
        
decoded_message = reversed_word1 + " " + shifted_word2
print(f"Decoded Message: {decoded_message}")        
```
