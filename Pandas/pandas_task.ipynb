## Assignment 2: 
  ### The null_handling_strategy Function


#### Scenario: Incoming user data often has missing values.Students must implement a flexible strategy to handle these "Null Values" to prepare data for Machine Learning.
### Requirements:

* Check for null values in the DataFrame.

* Apply a strategy based on parameters: "drop_rows", "fill_mean", or "fill_median" .

* Ensure the function only fills numerical columns when using mean or median.

```python
def null_handling_strategy(df, strategy="fill_mean"):
    copy_df = df.copy()

    has_nulls = copy_df.isnull().values.any()
    if not has_nulls:
        return copy_df

    if strategy == "drop_rows":
        copy_df = copy_df.dropna()

    elif strategy in ["fill_mean", "fill_median"]:
        numeric_cols = [col for col in copy_df.columns if is_numeric_dtype(copy_df[col])]

        for col in numeric_cols:
            if copy_df[col].isnull().any():
                if strategy == "fill_mean":
                    fill_value = copy_df[col].mean()
                else:
                    fill_value = copy_df[col].median()

              
                copy_df[col] = copy_df[col].fillna(fill_value)

    else:
        raise ValueError(f"Unknown strategy '{strategy}'. Use 'drop_rows', 'fill_mean', or 'fill_median'.")

    return copy_df

null_handling_strategy(df_test, strategy="fill_mean")
```
